```python
with open('p1.txt', 'r') as f:
    lines = f.readlines()
```

```python
dir = {}
current = []

for line in lines:
    if line[0] == '$':
        if line.strip() != '$ ls':
            command = [each.strip() for each in line.split(' ')]
            if command[2] == '..':
                current.pop()
                continue

            if command[2] == '/':
                current = ['.']
            else:
                current.append(command[2])
                
            path = '/'.join(current)
            if not(dir.get(path)):
                dir[path] = 0

    else:
        if line[:3] != 'dir':
            size = int(line.split()[0])
            for i in range(len(current)):
                path = '/'.join(current[:i+1])
                dir[path] += size
```

```python
total_space = 70000000
need = 30000000
need_to_free = need + dir['.'] - total_space
```

```python
ans = 70000000
for key, val in dir.items():
    if val >= need_to_free and val<ans:
        ans = val
print(ans)
```
